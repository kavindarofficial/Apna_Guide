```python
dfd1=df.dropna(axis=0)
dfd1.columns
```

```python
print(len(df),len(dfd1))
```

```python
dfd1=dfd1.drop(['Date','Location'],axis=1)
```

```python
data=pd.get_dummies(dfd1)
print(len(data.columns))
data.head()
```

```python
x=data.drop(['RainTomorrow_Yes','RainTomorrow_No'],axis=1)
y=pd.DataFrame(data['RainTomorrow_Yes'])
y['RainTomorrow_Yes']=y['RainTomorrow_Yes'].apply(lambda x: 1 if x else 0)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
pred=model.predict(x_test)
```

```python
print('Accuracy : ',accuracy_score(y_test,pred))
```
