```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import re
from sklearn.pipeline import Pipeline
```

```python
from google.colab import drive
drive.mount('/content/drive')
test_path="/content/drive/MyDrive/Colab_Notebooks/Disaster_or_Not/test.csv"
train_path="/content/drive/MyDrive/Colab_Notebooks/Disaster_or_Not/train.csv"
```

```python
df=pd.read_csv(train_path)
test_df=pd.read_csv(test_path)
df.head()
```

```python
len(df)
```

```python
unique_keywords=df['keyword'].unique()
len(unique_keywords)
```

```python
unique_places=df['place'].unique()
len(unique_places)
```

```python
df['keyword'].fillna('unknown', inplace=True)
df['place'].fillna('unknown', inplace=True)

test_df['keyword'].fillna('unknown', inplace=True)
test_df['place'].fillna('unknown', inplace=True)

df['text'] = df['keyword'] + ' ' + df['place'] + ' ' + df['tweet']
test_df['text']=test_df['keyword']+ ' ' + test_df['place'] + ' ' + test_df['tweet']

X = df['text']
y = df['disaster']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

pipeline = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('clf', LogisticRegression())
])

pipeline.fit(X_train, y_train)
```

```python
y_pred = pipeline.predict(X_test)
print(f'Accuracy : ',accuracy_score(y_test, y_pred))
```
